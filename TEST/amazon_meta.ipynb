```python
# import sys
import time

input_path = './dataset/Amazon/amazon-meta.txt' #sys.argv[1]

meta_f = open(input_path, 'r', encoding="UTF-8")
id_dict = {}
id_list = []
item_list = []
count = 0
id_list.append('0')
item_list.append('None')
while True:
    line = meta_f.readline().replace('\n', '')
    
    if 'Id:' in line:
        line = line.split()
        if line[1] == '0':
            continue
        id_list.append(line[1])
        c = 1
        
    if 'group:' in line:
        line = line.split()
        if c == 1:
            item_list.append(line[1])
            c = 0
        elif c == 0:
            print(id_list)
            item_list.append('None')
            item_list.append(line[1])
    
    count = count + 1
    
    # print(c)
    if count == 15010575:
        break


meta_f.close
```
