```python
queryList3 = []
f = open(randompath, 'r')
while True:
    line = f.readline().replace('\n', '').strip()
    if not line:
        break
    v = line.split(Separator)[0]
    u = line.split(Separator)[1]
    queryList3.append([v, u])

f.close
print('질의3 개수')
print(len(queryList3))

s1 = time.time()
Correct = 0
wrong= 0
wronglist =[]

# print(queryList3)
def BFS(sq, eq):
    visited = []
    if sq in ptn_pt2:
        sq = ptn_pt2[sq]
    
    if eq in ptn_pt2:
        eq = ptn_pt2[eq]
        
    queue = [sq]
    
    t = 0
    while queue:
        t += 1
        n = queue[0]
        del queue[0]
        
    
                
        if n == eq:
            return True
        
        # if t == 100:
        #     return False
        sq_n = list(gd.IG.vertex(n)[2].keys()) #이웃 정점 집합
        
        if n not in visited:
            visited.append(n)
            queue.extend(sq_n)
    
    return False


for sq, eq in queryList3:
    # print(f'시작: {sq} -> 도착: {eq}')
    if BFS(sq, eq) == True:
        Correct += 1
    else:
        wrong += 1
        
e = time.time() - s1
print(e, Correct, wrong, wronglist)
        
```
