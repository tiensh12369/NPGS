```python
from InitialGraphs import Graphdata
import sys
import time
import random

random.seed(100)

# fileName = 'DBLP-v1'
fileName = 'ENZYMES118'

in_path = './Result/FPGS2/' + fileName + '/' + fileName + '_2_20_0.edges'
in_label = './Result/FPGS2/' + fileName + '/' + fileName + '_2_20_0.node_labels'
in_pattern = './Result/FPGS2/' + fileName + '/' + fileName + '_2_20_0.pattern'
in_pattern_nodes = './Result/FPGS2/' + fileName + '/' + fileName + '_2_20_0.pattern_nodes'

Separator = ','
gd = Graphdata(in_path, in_label, Separator)

pattern = {} #{인코딩패턴, 패턴}
ptFile = open(in_pattern, 'r')

while True:
    line = ptFile.readline().replace('\n', '').strip()
    if not line:
        break
    
    pt = line.split(Separator)
    pattern[pt[1]] = pt[0]

ptFile.close

ptnodes = {} #{패턴넘버, 패턴노드}
ptnodes2 = {} #{패턴노드, 패턴 넘버}

ptnFile = open(in_pattern_nodes, 'r')
while True:
    line = ptnFile.readline().replace('\n', '').strip()
    if not line:
        break
    
    ptn = line.split(Separator)
    ptnodes[ptn[0]] = ptn[1]
    ptnodes2[ptn[1]] = ptn[0]

ptnFile.close

ptn_pt = {} #{인코딩패턴, 패턴노드}
# print(ptnodes.keys())
print(ptnodes)
print(ptnodes2)
for ptkey in ptnodes.keys():
    pt = ptkey.split('-')[0]
    if pt not in ptn_pt:
        ptn_pt[pt] = [ptnodes[ptkey]]
    ptn_pt[pt].append(ptnodes[ptkey])
print(len(ptn_pt))

elist = [] #edge list

for el1, el2 in gd.IG.edge_list:
    if el1 in ptnodes:
        el1 = ptnodes[el1]
    
    if el2 in ptnodes:
        el2 = ptnodes[el2]
    
    elist.append([el1, el2])
print(ptn_pt)

ptn_pt2 = {} #노드/패턴번호
for ptkey in ptnodes.keys():

    b = ptnodes[ptkey].split('/')
    for c in b:
        ptn_pt2[c] = ptkey
print(ptn_pt2)
```
