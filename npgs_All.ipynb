```python
from InitialGraphs import Graphdata
import NPGSAlgorithm as na
import os
import math
import time

def convert_size(size_bytes):
    if size_bytes == 0:
        return "0B"
    size_name = ("B", "KB", "MB", "GB", "TB", "PB", "EB", "ZB", "YB")
    i = int(math.floor(math.log(size_bytes, 1024)))
    p = math.pow(1024, i)
    s = round(size_bytes / p, 2)
    return "%s %s" % (s, size_name[i])

s1 = time.time()
fileName = 'DBLP-v1'

# for i in range(100):
#     IGPath = './Gen_Evolution_set/' + fileName + '/' + fileName + '_' + str(i) + '.edges'
#     lbPath = './dataset/' + fileName + '/' + fileName + '.node_labels'
    
IGPath = './dataset/' + fileName + '/' + fileName + '.edges' #sys.argv[1]
lbPath = './dataset/' + fileName + '/' + fileName + '.node_labels' #sys.argv[2]
Separator = ','
t = 2

G = Graphdata(IGPath, lbPath, Separator).IG
E = G.edge_list

Gc = na.Basic(G, t, E)
# print(len(E), len(Gc.edge_list))
sim_size = len(Gc.edge_list)
node_nume = len(Gc.node_dict)

savefile = './Result/NPGS/' + fileName + '_result_Basic.edges'
f = open(savefile, 'w')

for e in Gc.edge_list:
    str_e = e[0] + ',' + e[1] + '\n'
    f.write(str_e) 
f.close()

file_size = 0

file_size = file_size + os.path.getsize(savefile) 
print('OutPut File Size:', convert_size(file_size), 'bytes')

d1 = time.time()-s1
print(f'N: {node_nume}, E: {sim_size}, Time: {d1}')
```
